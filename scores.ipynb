```python
import json
```

```python
with open('output.jsonl', 'r') as json_file:
    out = list(json_file)
    
with open('yelp_challenge_8_with_answers.jsonl', 'r') as json_file:
    answers = list(json_file)
    
print(answers[1])
```

```python
acc = 0 
mae = 0
for i in range(len(out)):
    
    result = json.loads(out[i])
    ans = json.loads(answers[i])
    print("Res:", result['predicted_stars'],  "Ans:", ans['stars'], "Diff", result['predicted_stars'] -  ans['stars'])
    mae += abs(result['predicted_stars'] -  ans['stars'])
    if result['predicted_stars'] ==  ans['stars']:
        acc+=1
```

```python
print(len(out))
print("MAE:", abs(mae/len(out)))
print("ACC:", abs(acc/len(out)))
```
